### ASCII

ASCII, which stands for American Standard Code for Information Interchange, is a way to represent text using numbers. 

In ASCII, each character, like letters, numbers, and symbols, is assigned a unique number code. 

This allows computers to store and communicate text information in a digital format, making it easy to represent and process written language electronically.

### char() Function

The `char()` function in Python converts an integer (ASCII code) to its corresponding character. It takes an ASCII value as an argument and returns the character represented by that value.

```python
# Example of char() function
char_code = 65
character = chr(char_code)
print(character)  # Output: 'A'
```

### ord() Function

The `ord()` function in Python does the opposite of the `char()` function. It takes a character as an argument and returns its corresponding ASCII value (integer).


```python
# Example of ord() function
character = 'A'
char_code = ord(character)
print(char_code)  # Output: 65
```
